Pour savoir quelles stations sont les plus proches d'une métropole donnée, petit programme pour les afficher sur un fond de carte.

```python
pattern = r'(?P<d>[\d\.]+).*?(?P<m>[\d\.]+).*?(?P<s>[\d\.]+).*?(?P<h>[NSWE])'
hemisphere_sig = {'N': 1, 'S': -1, 'E': 1, 'W': -1}

dms = mfdata['LAT'].str.extract(pattern)
dms['h'] = dms['h'].map(hemisphere_sig)
dms = dms.astype(float)

mfdata['Lat'] = dms['h'] * (dms['d'] + dms['m'].div(60) + dms['s'].div(3600))

dms = mfdata['LON'].str.extract(pattern)
dms['h'] = dms['h'].map(hemisphere_sig)
dms = dms.astype(float)

mfdata['Lon'] = dms['h'] * (dms['d'] + dms['m'].div(60) + dms['s'].div(3600))
print(mfdata)

mfdata=mfdata.dropna(subset=['Lon'])

mfdata=mfdata.dropna(subset=['Lat'])

import folium
map_osm = folium.Map(location=[48.85, 2.34])
for ind, lat, lon, com in mfdata[['Lat', 'Lon', 'NOM']].itertuples():
    map_osm.add_child(folium.RegularPolygonMarker(location=[lat,lon], popup=com,
                       fill_color='#132b5e', radius=5))
map_osm
```

On génère une liste de dates du 1er janvier 2020 au 20 avril 2021.
